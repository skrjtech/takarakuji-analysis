```python
import re
import time
import random
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from tqdm.notebook import tqdm
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
chikan = re.compile('[()]')
```

```python
def startUp():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    service = Service(executable_path=ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    return driver
```

```python
TopUrl = "https://www.mizuhobank.co.jp"
```

```python
driver = startUp()
driver.get("https://www.mizuhobank.co.jp/retail/takarakuji/check/loto/backnumber/index.html")
html = driver.page_source.encode("utf-8")
soup = BeautifulSoup(html, "html.parser")
tables = soup.select("#mainCol > article > section.section.js-lottery-backnumber > div > table.typeTK.js-backnumber-b > tbody > tr > td:nth-child(1) > a")
lotos = []
for idx, table in enumerate(tables):
    url = TopUrl + table.get("href")
    driver.get(url)
    html = driver.page_source.encode("utf-8")
    soup = BeautifulSoup(html, "html.parser")
    tables_trs = soup.select("#mainCol > article > div > div > div.spTableScroll.sp-none > table > tbody > tr")
    for tr in tables_trs:
        count = tr.select_one("th").text
        vals = " ".join([t.text for t in tr.select("td")])
        lotos.append(f"{count} {vals}".split(" "))
    
    time.sleep(0.3)

lotosupdate = []
driver.get("https://www.mizuhobank.co.jp/retail/takarakuji/check/loto/backnumber/index.html")
html = driver.page_source.encode("utf-8")
soup = BeautifulSoup(html, "html.parser")
links = soup.select("#mainCol > article > section.section.js-lottery-backnumber > div > table:nth-child(2) > tbody > tr > td:nth-child(2) > a")[::-1]
for index, link in enumerate(links):
    url = TopUrl + link.get("href")
    driver.get(url)

    html = driver.page_source.encode("utf-8")
    soup = BeautifulSoup(html, "html.parser")
    tables = soup.select("#mainCol > article > section > section.section > section > div > div.sp-none > table")[::-1]

    html = driver.page_source.encode("utf-8")
    soup = BeautifulSoup(html, "html.parser")
    tables = soup.select("#mainCol > article > section > section.section > section > div > div.sp-none > table")[::-1]
    for table in tables:
        count = table.select_one("thead > tr > th.alnCenter.bgf7f7f7.js-lottery-issue-pc").text
        date = table.select_one("tbody > tr:nth-child(1) > td").text
        values = " ".join([ t.text for t in table.select("tbody > tr:nth-child(2) > td.alnCenter.extension > strong.js-lottery-number-pc")])
        bornas = table.select_one("tbody > tr:nth-child(2) > td.alnCenter.extension.green > strong.js-lottery-bonus-pc").text
        bornas = chikan.sub('', bornas)
        data = "{} {} {} {}".format(count, date, values, bornas).split(" ")
        lotosupdate.append(data)
        
    time.sleep(0.3)
    lotos.extend(lotosupdate)
    df = pd.DataFrame(lotos, columns="count date val1 val2 val3 val4 val5 bornas".split(' '))
    df = df.set_index("count")
    df.to_csv("miniloto.csv")
```

```python
driver = startUp()
TopUrl = "https://www.mizuhobank.co.jp/"
driver.get("https://www.mizuhobank.co.jp/retail/takarakuji/check/numbers/backnumber/index.html")
html = driver.page_source.encode("utf-8")
soup = BeautifulSoup(html, "html.parser")
links = soup.select("#mainCol > article > section > section:nth-child(2) > div > table.typeTK.js-backnumber-b > tbody > tr")
links = [link.get("href") for link in links for link in link.select("td > a")]

df_numbers3 = pd.DataFrame(columns=["count", "datetime", "val1", "val2", "val3"])
df_numbers4 = pd.DataFrame(columns=["count", "datetime", "val1", "val2", "val3", "val4"])
# for idx, link in enumerate(tqdm(links)):
#     flag = False
#     while True:
#         url = TopUrl + link
#         time.sleep(1.12 * random.random())
#         driver.get(url)
#         html = driver.page_source.encode("utf-8")
#         soup = BeautifulSoup(html, "html.parser")
#         tables_trs = soup.select("#mainCol > article > div > div > div.spTableScroll.sp-none > table > tbody > tr")
#         for tr in tables_trs:
#             count = tr.select_one("th.bgf7f7f7").text
#             if len(count) == 0:
#                 flag = True
#                 break
#             datetime = tr.select_one("td:nth-child(2)").text
#             data1 = [ count, datetime ]
#             data2 = [ count, datetime ]
#             for s in tr.select_one("td:nth-child(3)").text: data1.append(int(s))
#             for s in tr.select_one("td:nth-child(4)").text: data2.append(int(s))
#             df3 = pd.DataFrame([data1], columns=["count", "datetime", "val1", "val2", "val3"])
#             df4 = pd.DataFrame([data2], columns=["count", "datetime", "val1", "val2", "val3", "val4"])
#             df_numbers3 = pd.concat([df_numbers3, df3])
#             df_numbers4 = pd.concat([df_numbers4, df4])
#             flag = False
            
#         if flag:
#             print("idx: ", idx, " countinue ", "url: ", url)
#             continue
#         break
driver.get("https://www.mizuhobank.co.jp/retail/takarakuji/check/numbers/backnumber/index.html")
html = driver.page_source.encode("utf-8")
soup = BeautifulSoup(html, "html.parser")
numbers_links = [ soup.select("#mainCol > article > section > section:nth-child(2) > div > table:nth-child(2) > tbody > tr > td:nth-child(2) > a"),
                  soup.select("#mainCol > article > section > section:nth-child(2) > div > table:nth-child(2) > tbody > tr > td:nth-child(3) > a") ]

df_numbers = [df_numbers3, df_numbers4]
for idx, numbers_link in enumerate(tqdm(numbers_links, position=0)):
    for link in tqdm(numbers_link[::-1], position=1, leave=False):
        flag = False
        while True:
            url = TopUrl + link.get("href")
            time.sleep(0.3)
            driver.get(url)
            html = driver.page_source.encode("utf-8")
            soup = BeautifulSoup(html, "html.parser")
            tables = soup.select("#mainCol > article > section > section > section > div > div:nth-child(1) > table")[::-1]
            df = []
            for table in tables:
                count = table.select_one("thead > tr > th:nth-child(2)").text
                
                if len(count) == 0:
                    flag = True
                    break
                
                datetime = table.select_one("tbody > tr:nth-child(1) > td").text
                numbers = table.select_one("tbody > tr:nth-child(2) > td").text
                numbers = [int(i) for i in numbers]
                datas = [count, datetime] + numbers
                columns = ["count", "datetime"] + ["val1", "val2", "val3", "val4"][:len(numbers)]
                df.append(pd.DataFrame([datas], columns=columns))
                flag = False
            if flag:
                continue
            break
        df_numbers[idx] = pd.concat([df_numbers[idx], pd.concat(df)])
df_numbers3, df_numbers4 = df_numbers
```
